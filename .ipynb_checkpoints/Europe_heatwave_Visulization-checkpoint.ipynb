```python
import xarray as xr
import hvplot.xarray
import hvplot.pandas
import holoviews as hv

# 1) Initialize the extension (must run once per session)
hv.extension('bokeh')

# 2) Load data
p90 = xr.open_dataarray("H:/Climate/ERA5/Monthly_AVG_t2m_1940_2025/Clip/EU/p90_baseline_2000_2015_EU.nc")
freq = xr.open_dataarray("H:/Climate/ERA5/Monthly_AVG_t2m_1940_2025/Clip/EU/heatwave_frequency_2019_2024_EU.nc")
inten = xr.open_dataarray("H:/Climate/ERA5/Monthly_AVG_t2m_1940_2025/Clip/EU/heatwave_intensity_2019_2024_EU.nc")

```

```python
p90_map = p90.hvplot.image(
    x='longitude', y='latitude', cmap='coolwarm',
    title='P90 Baseline (°C) — Europe (2000–2015)',
    colorbar=True, clim=(10, 35),
    width=900, height=600,
    tools=['hover','pan','wheel_zoom','save']
)
p90_map
```

```python
p90_hist = p90.to_series().hvplot.hist(
    bins=50, width=900, height=500,
    xlabel='Temperature (°C)', ylabel='Cell count',
    title='Distribution of P90 Baseline (°C)',
    tools=['hover','pan','wheel_zoom','save']
)
p90_hist
```

```python
freq_map = freq.hvplot.image(
    x='longitude', y='latitude', cmap='Reds',
    title='Heatwave Frequency — Europe (2019–2024)',
    colorbar=True, clim=(0, 25),
    width=900, height=600,
    tools=['hover','pan','wheel_zoom','save']
)
freq_map
```

```python
freq_hist = freq.to_series().hvplot.hist(
    bins=72, width=900, height=500,
    xlabel='Months exceeding P90', ylabel='Cell count',
    title='Distribution of Heatwave Frequency (2019–2024)',
    tools=['hover','pan','wheel_zoom','save']
)
freq_hist
```

```python
inten_map = inten.hvplot.image(
    x='longitude', y='latitude', cmap='Oranges',
    title='Heatwave Intensity — Europe (2019–2024)',
    colorbar=True, clim=(0, 20),
    width=900, height=600,
    tools=['hover','pan','wheel_zoom','save']
)
inten_map
```

```python
inten_hist = inten.to_series().hvplot.hist(
    bins=100, width=900, height=500,
    xlabel='Sum exceedance (°C)', ylabel='Cell count',
    title='Distribution of Heatwave Intensity (2019–2024)',
    tools=['hover','pan','wheel_zoom','save']
)
inten_hist
```
